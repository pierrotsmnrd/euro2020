```python
from datetime import datetime
```

```python
# First URL
matches_france = download_url("https://match.uefa.com/v2/matches?language=FR&offset=0&limit=20&style=FULL&teamId=43")
```

```python
selection = ["2032240","2030818","2030791","2030762","2030108","2030080","2029167","2030063","2030032","2029150",]

focus = None
for m in matches_france:
    
    is_passed = datetime.strptime(m['kickOffTime']['date'], '%Y-%m-%d')  < datetime.now()
    
    #if (str(m['id']) in selection and not is_passed) or (str(m['id']) not in selection and is_passed ):
    #    print("WEIRD !")
    
    if str(m['id']) == '2032240':
        focus = m
        
    print( m['id'], "\tis passed" if is_passed else "\tis yet to come" ) 
    print( m['awayTeam']['countryCode'], m['homeTeam']['countryCode'] ) 
    print( m['competition']['code'] ) 
    print( m['competition']['id'] ) 
    print( m['competition']['metaData']['name'] ) 
    print( m['kickOffTime'] ) 
    print("")
```

```python
# have a second look, can extract interesting things from the raw match, like the referees for instance
pprint(focus)
```

```python
# ['count', 'result', 'limit', 'offset', 'tags', 'tookMs', 'aggregations']
pprint(match_data)
```
